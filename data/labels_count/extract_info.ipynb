The camera not used yet are Cameras 1 3 5 8 9 10 12 13 14 16 17 18. 
We are going to find through the labels count data, in which file we can found some triton crapauds grenouilles on our dataset, in order to create a new subset for the test.

```python
import pandas as pd
from datetime import datetime as dt
import numpy as np
```

```python
# Read the csv file
data = pd.read_csv('labels_Caméra01.csv')
data
```

```python
# Convert the timestamp column to datetime
data['dateheure'] = [dt.fromtimestamp(x) for x in data['timestamp']]

# Drop the timestamp column as we won't need it anymore
data = data.drop(columns=['timestamp'])
# Split datetime into date and time
data['date']   = [d.date() for d in data['dateheure']]
data['h_m_s']  = [d.time() for d in data['dateheure']]
data['h']      = [d.hour for d in data['dateheure']]

data['annee']  = [d.year for d in data['dateheure']]
data['mois']   = [d.strftime("%B") for d in data['dateheure']]
data['jour']   = [d.day for d in data['dateheure']]
# Les données obtenues ne sont pas tout à fait correctes, il faut considerer l'heure -1 . Parfois -2.
```

```python
# count the number of triton on the camera 1
data['triton'].where(data['triton'] == 1).value_counts()
```

```python
# retrieve the data of the triton
data_triton = data[data['triton'] == 1]
data_triton
```

```python
# let's  randomly take 10 tritons on this camera
tritons = data_triton.sample(n=20)
tritons
```

```python
# number of crapauds grenouilles on the camera 1
data['crapaud'].value_counts()
# There is no crapauds on the camera 1
```

```python
# number of grenouilles on the camera 1
data['grenouille'].value_counts()
```

```python
data_grenouille = data[data['grenouille'] != 0]
data_grenouille
```
