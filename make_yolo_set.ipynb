```python
import pandas as pd
import os

bb_csv = pd.read_csv("data/path_and_bounding_box.csv")
def from_class_to_yolo_cat(class_name):
    if class_name == "triton":
        return 1
    elif class_name == "grenouille-crapaud":
        return 2
    elif class_name == "planche":
        return 3
    elif class_name == "feuille":
        return 4
    elif class_name == "souris":
        return 5
    elif class_name == "insecte":
        return 6
    else:
        return -1
    
bb_csv['class'] = bb_csv['class'].transform(lambda x : int(from_class_to_yolo_cat(x)))
```

```python
subset_c1 = bb_csv[bb_csv['class'] == 1].iloc[:225,:]
subset_c2 = bb_csv[bb_csv['class'] == 2].iloc[:200,:]
subset_c4 = bb_csv[bb_csv['class'] == 4].iloc[:200,:]
subset_c5 = bb_csv[bb_csv['class'] == 5].iloc[:197,:]
subset_c1
#a = bb_csv[ bb_csv['path'] == 'camera_02_unnested/2017-03-13T21_12_07.jpg']
#a
```

```python
# creer un dossier labels
#labels = os.mkdir("data\labels")
def csv_to_yolo(subset) :
    for i in range(len(subset)) :
        name = subset.iloc[i,10].split(".")[0]
        file = open(r"data/labels/" + name + ".txt", "w")
        x_center = ((subset.iloc[i,1] + subset.iloc[i,3])/2)/subset.iloc[i,5]
        y_center = ((subset.iloc[i,2] + subset.iloc[i,4])/2)/subset.iloc[i,6]
        width = (subset.iloc[i,3] - subset.iloc[i,1])/subset.iloc[i,5]
        height = (subset.iloc[i,4] - subset.iloc[i,2])/subset.iloc[i,6]
        file.write(str(subset.iloc[i,0]) + " " + str(x_center) + " " + str(y_center) + " " + str(width) + " " + str(height))
        file.write("\n")
        file.close()
        
```

```python
# Afficher le contenu des fichiers txt contenus dans le dossier labels
csv_to_yolo(subset_c1)
```
